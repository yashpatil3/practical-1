```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
dt=pd.read_csv('Titanic.csv');
```

```python
dt
```

```python
dt.columns
```

```python
dt.drop(['PassengerId','Name','SibSp',
       'Parch','Ticket','Cabin','Embarked'], axis='columns' , inplace=True);
```

```python
dt
```

```python
dt.isnull().sum()
```

```python
age_mean = dt['Age'].mean()
```

```python
age_mean
```

```python
dt['Age'] = dt['Age'].fillna(age_mean)
```

```python
dt
```

```python
dt.isnull().sum()
```

```python
target=dt['Survived']
dt.drop(['Survived'],axis='columns',inplace=True);
```

```python
target
```

```python
dt
```

```python
dt1 = pd.get_dummies(dt, columns=['Sex'],dtype=int)
```

```python
dt1
```

```python
dt.columns
```

```python
x = dt1[['Pclass', 'Sex_female','Sex_male', 'Age', 'Fare']]
```

```python
x
```

```python
y=target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gs=GaussianNB()
```

```python
gs.fit(x_train, y_train)
```

```python
y_pred = gs.predict(x_test)
```

```python
from sklearn.metrics import precision_score,confusion_matrix,accuracy_score,recall_score
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
ps=precision_score(y_test,y_pred)
```

```python
ps
```

```python
cm=confusion_matrix(y_test,y_pred)
```
